```python

def Hour_Index(df):
    Index = []
    S = 0
    End = 0
    Hour_list = np.arange(1,25)
    for i in range(24):
        for j in range(len(df)):
            if df['Hour'][j]==Hour_list[i]:
                End = j
        Index.append((S,End))
        S = End+1
    return Index
```

```python
HIndex_S=Hour_Index(S_Data)
HIndex_D=Hour_Index(D_Data)
```

```python
def find_largest(IND):
    Res = IND[0][1]-IND[0][1]
    for i in range(len(IND)):
        if (IND[i][1]-IND[i][0])>=Res:
            Res = (IND[i][1]-IND[i][0])
    return Res

HIndex_D[1]
```

```python
Res = []    
def comb(Index,df,H,Res):
    size = find_largest(Index)
    for i in range(size+1):
        if df['price'][HIndex_D[H-1][0]+i] == df['price'][HIndex_D[H][0]+i]:
            Res.append((df['price'][HIndex_D[H-1][0]+i],df['Quantity'][HIndex_D[H-1][0]+i]+df['Quantity'][HIndex_D[H][0]+i]))
        else:
            mprice = np.argmin([df['price'][HIndex_D[H-1][0]+i],df['price'][HIndex_D[H][0]+i]])
            if mprice == 0:
                MH = H-1
                MaxH =H
            else:
                MH = H
                MaxH=H-1
            grad = (df['Quantity'][HIndex_D[MaxH][0]+i]- df['Quantity'][HIndex_D[MaxH][0]+i-1])/(df['price'][HIndex_D[MaxH][0]+i]- df['price'][HIndex_D[MaxH][0]+i-1])  
            x = df['price'][HIndex_D[MaxH][0]+i]*grad+df['Quantity'][HIndex_D[MH][0]+i-1]-df['price'][HIndex_D[MH][0]+i-1]*grad
            ResQ = df['Quantity'][HIndex_D[MH][0]+i] + x
            Res.append((df['price'][HIndex_D[MH][0]+i],ResQ))                            
            
```

```python
comb(HIndex_D,D_Data,1,Res)
```
